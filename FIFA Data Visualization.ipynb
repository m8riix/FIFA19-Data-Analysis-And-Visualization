```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
#import data
fifa_data=pd.read_csv("C:\\Users\\Asus1\\Desktop\\fifa19.csv")
fifa_data.head(10)
```

```python
# checking if the data contains any NULL value
#fifa_data.isnull().sum()
```

```python
# filling the missing value for proper data visualization (data cleaning)
fifa_data['Club'].fillna('No Club',inplace=True)
fifa_data['Work Rate'].fillna('Medium/ Medium', inplace = True)
fifa_data['Skill Moves'].fillna(fifa_data['Skill Moves'].median(), inplace = True)
fifa_data['Weak Foot'].fillna(3, inplace = True)
fifa_data['Preferred Foot'].fillna('Right', inplace = True)
fifa_data['International Reputation'].fillna(fifa_data['International Reputation'].mean(), inplace = True)
fifa_data['Wage'].fillna('€200K', inplace = True)
fifa_data['Weight'].fillna('200lbs', inplace = True)
fifa_data['Contract Valid Until'].fillna(2019, inplace = True)
fifa_data['Height'].fillna("5'11", inplace = True)
fifa_data['Loaned From'].fillna('None', inplace = True)
fifa_data['Joined'].fillna('Jul 1, 2018', inplace = True)
fifa_data['Jersey Number'].fillna(8, inplace = True)
fifa_data['Body Type'].fillna('Normal', inplace = True)
fifa_data['Position'].fillna('ST', inplace = True)
fifa_data['ShortPassing'].fillna(fifa_data['ShortPassing'].mean(), inplace = True)
fifa_data['Volleys'].fillna(fifa_data['Volleys'].mean(), inplace = True)
fifa_data['Dribbling'].fillna(fifa_data['Dribbling'].mean(), inplace = True)
fifa_data['Curve'].fillna(fifa_data['Curve'].mean(), inplace = True)
fifa_data['FKAccuracy'].fillna(fifa_data['FKAccuracy'], inplace = True)
fifa_data['LongPassing'].fillna(fifa_data['LongPassing'].mean(), inplace = True)
fifa_data['BallControl'].fillna(fifa_data['BallControl'].mean(), inplace = True)
fifa_data['HeadingAccuracy'].fillna(fifa_data['HeadingAccuracy'].mean(), inplace = True)
fifa_data['Finishing'].fillna(fifa_data['Finishing'].mean(), inplace = True)
fifa_data['Crossing'].fillna(fifa_data['Crossing'].mean(), inplace = True)
```

```python
# defining a function for cleaning the Weight data and applying the function to weight column
def lbs_to_flt(value):
  val = value.replace('lbs', '')
  return float(val)

fifa_data['Weight'] = fifa_data['Weight'].apply(lambda x : lbs_to_flt(x))
```

```python
# defining a function for cleaning the value and wage data and applying the function to respect columns
def str_removal(Value):
    out = Value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K', ''))*1000
    return float(out)
fifa_data['Value'] = fifa_data['Value'].apply(lambda x: str_removal(x))
fifa_data['Wage'] = fifa_data['Wage'].apply(lambda x: str_removal(x))
```

```python
# top wages of players and their respective names
top_wages_of_players=fifa_data.sort_values('Wage',ascending=False)
top_wages_of_players_list=list(top_wages_of_players['Wage'].head(15))
top_players_list=list(top_wages_of_players['Name'].head(15))
```

```python
#plotting bar graph for top 20 players according to their wages
#plotting bar graph for top 20 players according to their wages
plt.figure(figsize=(16,9))
plt.bar(top_players_list[::-1],top_wages_of_players_list[::-1],edgecolor='m',color='m',width=0.8)
plt.title("Top 15 FIFA Players According To Their Wages",color='y',fontsize=15)
plt.xlabel("Players Name",fontsize=17,color='r')
plt.ylabel("Wages (values in Euro)",fontsize=12,color='r')
plt.xticks(rotation=30)
```

```python
# Height of Players

plt.figure(figsize = (16, 9))
ax = sns.countplot(x = 'Height', data = fifa_data, palette = 'dark')
ax.set_title(label = 'Count of players on Basis of Height', fontsize = 20)
ax.set_xlabel(xlabel = 'Height in Foot per inch', fontsize = 16)
ax.set_ylabel(ylabel = 'Count', fontsize = 16)
plt.xticks(rotation=30)
plt.show()
```

```python
# To show Different body weight of the players participating in the FIFA 2019

plt.figure(figsize = (16, 9))
plt.style.use('_classic_test')
sns.distplot(fifa_data['Weight'], color = 'pink')
plt.title('Different Weights of the Players Participating in FIFA 2019', fontsize = 20)
plt.xlabel('Heights associated with the players', fontsize = 16)
plt.ylabel('count of Players', fontsize = 16)
plt.show()
```

```python
# To show Different Speciality Score of the players participating in the FIFA 2019

x = fifa_data.Special
plt.figure(figsize = (16,9))
plt.style.use('tableau-colorblind10')

ax = sns.distplot(x, bins = 58, kde = False, color = 'm')
ax.set_xlabel(xlabel = 'Special score range', fontsize = 16)
ax.set_ylabel(ylabel = 'Count of the Players',fontsize = 16)
ax.set_title(label = 'Histogram for the Speciality Scores of the Players', fontsize = 20)
plt.show()
```

```python
# To show Different overall scores of the players participating in the FIFA 2019

sns.set(style = "dark", palette = "deep", color_codes = True)
x = fifa_data.Overall
plt.figure(figsize = (12,8))
plt.style.use('ggplot')

ax = sns.distplot(x, bins = 52, kde = False, color = 'r')
ax.set_xlabel(xlabel = "Player\'s Scores", fontsize = 16)
ax.set_ylabel(ylabel = 'Number of players', fontsize = 16)
ax.set_title(label = 'Histogram of players Overall Scores', fontsize = 20)
plt.show()
```
